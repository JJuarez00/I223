```python
# 1. Generate a random integer. Show 2 methods!
# ----------------------------------------------- #

# Method 1 #
import random
one = []

# Prints 10 random
for i in range(10):
    num = random.randint(-10, 10) # Returns any random integer from -10 to 10
    one.append(num)
print(one)

# ----------------------------------------------- #

# Method 2 #

two = []

for i in range(10):
    num = random.randrange(-10, 11) # Returns a random integer from -10 to 10 (used -11 since n-1)
    two.append(num)
print(two)

```

```python
# 2. Guessing Game.

# Generate a random number between 1 and 9 (including 1 and 9). 
# Ask the user to guess the number, then tell them whether they guessed too low, too high, or exactly right.
# (Hint: remember to use the user input from the lecture 7 exercise).

# Extras:
# - Keep the game going until the user types “exit”
# - Keep track of how many guesses the user has taken, and when the game ends, print this out.

import random

randNum = random.randint(1, 9) # Returns any random integer from 1 to 9

def guess(a, b):
    if a == b:
        return "exit"
    elif a > b:
        return "too high"
    elif a < b:
        return "too low" 

tries = 0
while True:
    try:
        userNum = input("Guess the number (from 1 to 9) [type `exit` to quit]: ")

        if userNum == "exit":
            break

        num = int(userNum) # Make userNum into int, if it can be converted then throw ValueError.
        if num <= 0 or num >= 10:
                print(f"Make sure your value is in the range of 1 through 9.\tValue Entered: {userNum}")
        else:
            token = guess(num, randNum)
            if token == "exit":
                tries += 1
                print(f"\nCorrect!\nValue Entered: {userNum}\n# of Tries: {tries}")
                break
            elif token == "too high":
                print(f"Too High \t Value Entered: {userNum}")
                tries += 1
            elif token == "too low":
                print(f"Too Low \t Value Entered: {userNum}")
                tries += 1
            else:
                break

    except ValueError:
            print("\nError, invalid input.\nMake sure the input value is an integer!")
            tries += 1
```

```python
# 3. Rock Paper Scissors

# Make a two-player Rock-Paper-Scissors game. 
#       (Hint: Ask for player plays (using input ),
#        compare them, print out a message of congratulations to 
#        the winner, and ask if the players want to start a new game)

# Remember the rules:
#   - Rock beats scissors
#   - Scissors beats paper
#   - Paper beats rock

o = ["rock", "paper", "scissors"]


def rules(a, b):

    print(f"P1: {a}")
    print(f"P2: {b}")

    if a==b:
        print("Tie")

        # 0 = rock
        # 1 = paper
        # 2 = scissors

    elif (a == o[0] and b == o[2]) or \
         (a == o[2] and b == o[1]) or \
         (a == o[1] and b == o[0]):
        print(f"{a} beats {b}\nPlayer 1 Wins!\n")
    else:
        print(f"{b} beats {a} \nplayer 2 Wins!\n")

try:
    while True:
        while True:
            p1 = str(input("Player 1: enter Rock, Paper, or Scissors: ")).strip().lower()
            if p1 in o:
                break
            else:
                print("input has to match: `Rock`, `Paper`, or `Scissors`") 
        
        while True:           
            p2 = str(input("Player 2: enter Rock, Paper, or Scissors: ")).strip().lower()
            if p2 in o:
                break
            else:
                print("input has to match: `Rock`, `Paper`, or `Scissors`")
        
        rules(p1, p2)

        again = int(input("Play Again?\n(1 = YES and 0 = NO)"))
        if again == 1:
            continue
        elif again == 0:
            break
        else:
            print("hmmm, interesting. I take that as a YES!") # If user does not follow the (1 or 0) rule then play again lol.

except ValueError:
    print("Invalid Input! Try again!")
```
